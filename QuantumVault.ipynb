## 1. Resumen Ejecutivo

QuantumVault S.L. es una startup tecnológica especializada en almacenamiento cloud seguro mediante criptografía post-cuántica y arquitectura zero-knowledge.

La empresa ofrece un servicio similar a Google Drive, pero diseñado para resistir amenazas futuras derivadas del desarrollo de ordenadores cuánticos, utilizando algoritmos criptográficos post-cuánticos estandarizados por el NIST.

El proyecto se evaluará mediante técnicas clásicas de evaluación de inversiones: flujo de fondos, VAN, TIR, Payback, análisis de sensibilidad y análisis de riesgo.

# 20. Costo de capital (WACC)

Para evaluar financieramente el proyecto se emplea el coste medio ponderado del capital (WACC).

---

## 20.1 Coste del capital propio (Ke)

Dado el elevado riesgo asociado a startups tecnológicas, se adopta una rentabilidad mínima exigida del:

Ke = 12 %

---

## 20.2 Coste de la deuda (Kd)

Tipo de interés del préstamo:

Kd = 5 %

Tras impuestos:

Kd × (1 − T)

5 % × (1 − 0,25)

= 3,75 %

---

## 20.3 Cálculo del WACC

Fórmula:

WACC = E/V × Ke + D/V × Kd × (1 − T)

Sustituyendo:

WACC = 0,40 × 12 % + 0,60 × 5 % × (1 − 0,25)

WACC = 4,80 % + 2,25 %

WACC = 7,05 %

Por prudencia financiera se utilizará una tasa de descuento del:

8 %

en el cálculo del VAN.

## 24. Tratamiento fiscal

A diferencia de ciertos modelos utilizados en otros países, en España el Impuesto sobre Sociedades no grava directamente la facturación ni el beneficio bruto.

El impuesto se calcula sobre la base imponible, aproximada en este estudio mediante el beneficio antes de impuestos (BAI):

BAI = Ingresos − Gastos deducibles − Amortizaciones − Intereses

Impuesto:

IS = max(0, BAI × 25%)

Por tanto, las pérdidas no generan impuesto.

---

## 25. Flujo de Fondo Puro (Proyecto)

El flujo de fondo puro mide la rentabilidad intrínseca del proyecto independientemente de su financiación.

Flujo de caja libre:

FCL = EBIT × (1 − T) + Amortización − CAPEX − Δ Capital de trabajo

Se utilizará para calcular:

- VAN del proyecto.
- TIR del proyecto.

---

## 26. Flujo de Fondo del Inversionista

El flujo del inversionista incorpora el efecto de la deuda.

FF Inversionista = Flujo de Fondo Puro − Servicio de deuda

Este flujo permite evaluar la rentabilidad para los accionistas.

Se calcularán:

- VAN del inversionista.
- TIR del inversionista.

Siguiendo la metodología utilizada en AGROGAN, la evaluación del proyecto se realizará mediante flujos de fondos en términos reales.

Los flujos reales eliminan el efecto de la inflación y permiten evaluar la rentabilidad económica del proyecto en poder adquisitivo constante.

Por coherencia metodológica, la tasa de descuento utilizada en el cálculo del VAN será una tasa real.

En caso de utilizar tasas nominales de mercado, estas serán transformadas a tasas reales mediante la ecuación de Fisher:

(1 + i_nominal) = (1 + i_real) × (1 + inflación)

Todas las proyecciones financieras del modelo se expresarán en euros constantes del año 2026.
